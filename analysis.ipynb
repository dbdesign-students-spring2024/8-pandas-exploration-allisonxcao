### Handle Missing Values

```python
data['School Name'].fillna('Unknown', inplace=True)
data.sample(5)
```

### Convert Data Types

```python
data['Mean Scale Score'] = pd.to_numeric(data['Mean Scale Score'], errors='coerce')
data.sample(5)
```

### Check Supressed Data

```python
columns_to_check = ['Num Level 1', 'Pct Level 1', 'Num Level 2', 'Pct Level 2',
                    'Num Level 3', 'Pct Level 3', 'Num Level 4', 'Pct Level 4']
for column in columns_to_check:
    data[column] = data[column].replace('s', pd.NA)

data.sample(5)
```

### Remove Duplicates

```python
data.drop_duplicates(inplace=True)
display(data.sample(5))
```

### Mean, Min, and Max scores by Grade Level
This calculates data by grade to provide insights into performance levels across different grades

```python
data['Mean Scale Score'] = pd.to_numeric(data['Mean Scale Score'], errors='coerce')
grade_stats = data.groupby('Grade')['Mean Scale Score'].agg(['mean', 'min', 'max']).sort_values('mean', ascending=False)
grade_stats
```

### Performance of student based on category
This cell calculates average scores by student categories to identify disparities amongst test takers. 

```python
category_performance = data.groupby('Student Category')['Mean Scale Score'].agg(['mean', 'min', 'max']).sort_values('mean', ascending=False)
category_performance
```

### Yearly Average Performance
The average mean score per year

```python
data['Year'] = data['Year'].astype(int)
yearly_performance = data.groupby('Year')['Mean Scale Score'].mean()
yearly_performance = yearly_performance.sort_index()
yearly_performance
```

### Number of students tested by year and grade
The cell shows the number of students tested per year which can help with understanding test participation trends over time. 

```python
students_tested_summary = data.groupby(['Year', 'Grade'])['Number Tested'].sum()
data['Year'] = data['Year'].astype(int)  # Ensure 'Year' is integer
data['Grade'] = data['Grade'].astype(str)  # Ensure 'Grade' is string for correct grouping
students_tested_summary = data.groupby(['Year', 'Grade'])['Number Tested'].sum().unstack(level=1)
students_tested_summary
```

### Performance Level Distribution by Grade
The cell calculates the percentage distribution of each performance level across grades which demonstrates academic achievement levels across grades. 

```python
data['Num Level 1'] = pd.to_numeric(data['Num Level 1'], errors='coerce')
data['Num Level 2'] = pd.to_numeric(data['Num Level 2'], errors='coerce')
data['Num Level 3'] = pd.to_numeric(data['Num Level 3'], errors='coerce')
data['Num Level 4'] = pd.to_numeric(data['Num Level 4'], errors='coerce')
total_students_by_grade = data.groupby('Grade')['Number Tested'].sum()
level_counts_by_grade = data.groupby('Grade')[['Num Level 1', 'Num Level 2', 'Num Level 3', 'Num Level 4']].sum()
level_percentages_by_grade = level_counts_by_grade.div(total_students_by_grade, axis=0) * 100
level_percentages_by_grade
```

### Distribution of Mean Scale Scores
The histogram depicts the distribution of mean scale scores across all tests. 

```python
data['Mean Scale Score'] = pd.to_numeric(data['Mean Scale Score'], errors='coerce')
plt.figure(figsize=(10, 6))
plt.hist(data['Mean Scale Score'].dropna(), bins=30, color='blue', alpha=0.7)
plt.title('Distribution of Mean Scale Scores')
plt.xlabel('Mean Scale Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Yearly Trends in Mean Scores
The line graph depicts the yearly trends in average mean scores and how they changed over time

```python
yearly_scores = data.groupby('Year')['Mean Scale Score'].mean()
plt.figure(figsize=(10, 6))
yearly_scores.plot(kind='line', marker='o', color='green')
plt.title('Yearly Trends in Mean Scale Scores')
plt.xlabel('Year')
plt.ylabel('Average Mean Scale Score')
plt.grid(True)
plt.show()
```

### Performance by Student Category
The bar chart compares the average mean scale scores across different student categories which can help mitigate educational disparities. 

```python
category_scores = data.groupby('Student Category')['Mean Scale Score'].mean().sort_values()
plt.figure(figsize=(12, 8))
category_scores.plot(kind='barh', color='magenta')
plt.title('Average Mean Scale')
```

### Performance Levels Across Grades
The bar chart shows the distribution of performance levels across different grades. 

```python
data['Num Level 1'] = pd.to_numeric(data['Num Level 1'], errors='coerce')
data['Num Level 2'] = pd.to_numeric(data['Num Level 2'], errors='coerce')
data['Num Level 3'] = pd.to_numeric(data['Num Level 3'], errors='coerce')
data['Num Level 4'] = pd.to_numeric(data['Num Level 4'], errors='coerce')
levels_by_grade = data.groupby('Grade')[['Num Level 1', 'Num Level 2', 'Num Level 3', 'Num Level 4']].sum()
plt.figure(figsize=(12, 8))
levels_by_grade.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Performance Levels Across Grades')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.legend(title='Performance Level')
plt.show()
```

### Grade Level Performance over Time
The line chart displays the average mean scale score over time across all grades

```python
data['Year'] = data['Year'].astype(int)
data['Mean Scale Score'] = pd.to_numeric(data['Mean Scale Score'], errors='coerce')
grade_yearly_performance = data.groupby(['Year', 'Grade'])['Mean Scale Score'].mean().unstack()
plt.figure(figsize=(14, 8))
grade_yearly_performance.plot(kind='line', marker='o')
plt.title('Grade-Level Performance Over Time')
plt.xlabel('Year')
plt.ylabel('Average Mean Scale Score')
plt.legend(title='Grade')
plt.grid(True)
plt.show()
```
